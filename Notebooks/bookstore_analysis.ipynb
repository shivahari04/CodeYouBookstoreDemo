```python
# load in libraries
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

### Query 1: List the title, genre, and full author name for every book sold at a store in the Pacific region. Show each sale as its own row 

```python
query1 = """
SELECT 
b.title, g.name as genre,
a.first_name || '' || a.last_name AS author_name
FROM transactions t
JOIN stores s ON s.store_id = t.store_id
JOIN books b ON t.book_id = b.book_id
JOIN genres g on g.genre_id = b.genre_id
JOIN authors a ON a.author_id = b.author_id
WHERE s.store_region ='Pacific'

 """
```

```python
# execute the query
result1 = pd.read_sql_query(query1,connection)
result1
```

```python
# create a visualization
genre_counts = result1['genre'].value_counts()

plt.figure(figsize=(10, 6))
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

genre_counts.plot(kind='bar', ax=ax)

plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Pacific Sales by Genre')
plt.tight_layout()
plt.show()
```
